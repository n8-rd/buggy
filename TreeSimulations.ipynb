The simplest models of phylogenetic branching are the ERM models = Equal Rates Markovian models. The simplest of these is the pure-birth model, a.k.a, the Yule model. It has one paramter -- the birth rate = the speciation rate = lambda.


First we need to load some R libraries. These are chunks of code that expand the core functionality of R.

```r
library(phytools, ape)
```

Now we can do some simple Yule simulations.

```r
print('doing Yule simulations...') 
pbs.4 <- pbtree(b=0.4, d=0, t=5, nsim=4)
pbs.9 <- pbtree(b=0.9, d=0, t=5, nsim=4)
```

Here we are using the pbtree (short for pure-birth tree) function from the R package phytools. In each call to this function we pass four arguments: 1) b is the birth rate, that is, the expected number of speciation events per unit time; 2) d is the death rate, that is, the extinction rate; 3) t is for time, that is, how many time steps we want the simulation to run, and 4) nsim is short for number of simulations. Note that we've set extinction rates to zero. That's what makes this a Yule model. Also note that in our first call of the function, our speciation rate was a little less than half of what it was in our second call.

OK. Let's look at some trees!

```r
#first look at the four trees with simulated with the birth rate at 0.4.
par(mfrow=c(2,2))
for (t in pbs.4){
    plot(t)
    }
```

```r
print('doing birth-death simulations...') 
bds.4 <- pbtree(b=0.4, d=0.2, t=5, nsim=4)
```
